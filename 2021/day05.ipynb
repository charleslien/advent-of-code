```python
vals = []
for line in lines:
  (a, b), (c, d) = [x.split(',') for x in line.split(' -> ')]
  vals.append([int(x) for x in [a, b, c, d]])
```

```python
counts = defaultdict(lambda: 0)
for a, b, c, d in vals:
  if a == c:
    for y in range(min(b, d), max(b, d) + 1):
      counts[a, y] += 1
    continue
  if b == d:
    for x in range(min(a, c), max(a, c) + 1):
      counts[x, b] += 1
    continue
np.sum([x > 1 for x in counts.values()])
```
