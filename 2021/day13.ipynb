```python
coords = [[int(x) for x in line.split(',')] for line in blocks[0]]
p = re.compile(r'fold along (\w)=(\d+)')
inst = [p.match(line).groups() for line in blocks[1]]
inst = [[x[0], int(x[1])] for x in inst]
```

```python
def fold(coords, line):
  r = set()
  for c in coords:
    if line[0] == 'x':
      r.add((line[1] - abs(c[0] - line[1]), c[1]))
    else:
      r.add((c[0], line[1] - abs(c[1] - line[1])))
  return r
```

```python
c = coords
for i in inst:
  c = fold(c, i)
```

```python
ca = np.array(list(c))
minx, miny = np.min(ca, axis=0)
maxx, maxy = np.max(ca, axis=0)
```

```python
for y in range(miny, maxy + 1):
  line = []
  for x in range(minx, maxx + 1):
    if (x, y) in c:
      line.append('#')
    else:
      line.append('.')
  print(''.join(line))
```
