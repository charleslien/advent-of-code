```python
blocks = process_raw(raw)
```

```python
scanners = []
for block in blocks:
  scanners.append([])
  for line in block[1:]:
    scanners[-1].append([int(x) for x in line.split(',')])
  scanners[-1] = np.array(scanners[-1])
```

```python
def get_overlaps(s1, s2):
  set1 = s1 if isinstance(s1, set) else set(tuple(x) for x in s1)
  set2 = set(tuple(x) for x in s2)
  return len(set1 & set2)
```

```python
def get_max_trans_overlaps(s1, s2):
  mov = 0, None
  for i, row1 in enumerate(s1):
    for j, row2 in enumerate(s2):
      trans2 = s2 - row2 + row1
      ov = get_overlaps(s1, trans2)
      if ov > mov[0]:
        mov = ov, trans2
        if mov[0] >= 12:
          return mov
  return mov
```

```python
def get_max_overlaps(s1, s2):
  mov = 0, None
  for p in permutations([0, 1, 2]):
    for flip in powerset(p):
      flipped2 = np.array(s2)
      flipped2[:, flip] = -flipped2[:, flip]
      flipped2 = flipped2[:, p]
      trans_overlaps = get_max_trans_overlaps(s1, flipped2)
      if trans_overlaps[0] > mov[0]:
        mov = trans_overlaps
        if mov[0] >= 12:
          return mov
  return mov
```

```python
# Takes around 5 min...
to_fix = set(range(1, len(scanners)))
fixed = {tuple(x) for x in scanners[0]}
while to_fix:
  for i in range(len(scanners)):
    if i not in to_fix:
      continue
    mov = get_max_overlaps(fixed, scanners[i])
    if mov[0] >= 12:
      fixed |= {tuple(x) for x in mov[1]}
      to_fix.remove(i)
```
