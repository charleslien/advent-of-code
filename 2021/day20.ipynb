```python
ref = blocks[0][0]
screen = np.zeros([len(blocks[1]), len(blocks[1][0])], dtype=int)
for i, line in enumerate(blocks[1]):
  for j, p in enumerate(line):
    if p == '#':
      screen[i, j] = 1
```

```python
def expand(screen, margin, extra):
  out = np.zeros(np.array(np.shape(screen)) + margin * 2, dtype=int)
  for a, b in np.transpose(np.where(screen == margin)) + 1:
    out[a, b] = 1
  out[:margin, :] = extra
  out[-margin:, :] = extra
  out[:, :margin] = extra
  out[:, -margin:] = extra
  return out
```

```python
extra = 0
nscreen = expand(screen, 1, extra)
```

```python
def get_output(screen, extra):
  inp = expand(screen, 1, extra)
  if ref[int_round(Polynomial([extra] * 9)(2))] == '#':
    extra = 1
  else:
    extra = 0
  out = np.zeros_like(inp)
  out[0, :] = extra
  out[-1, :] = extra
  out[:, 0] = extra
  out[:, -1] = extra
  n, m = np.shape(inp)
  for i in range(1, n - 1):
    for j in range(1, m - 1):
      window = inp[i-1:i+2, j-1:j+2]
      index = int_round(Polynomial(window.flatten()[::-1])(2))
      if ref[index] == '#':
        out[i, j] = 1
  return out, extra
for _ in range(2):
  nscreen, extra = get_output(nscreen, extra)
np.sum(nscreen)
```
