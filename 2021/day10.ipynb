```python
total = 0
for line in lines:
  stack = []
  for c in line:
    if c in ')>]}':
      if not stack:
        total += points[c]
        break
      prev = stack.pop()
      if prev + c not in ['()', '<>', '[]', '{}']:
        total += points[c]
        break
    else:
      stack.append(c)
total
```
