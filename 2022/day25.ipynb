```python
def to_num(s):
  total = 0
  for d in s:
    total *= 5
    if d in '012':
      total += int(d)
    if d == '-':
      total -= 1
    if d == '=':
      total -= 2
  return total
```

```python
def to_snafu(n):
  digits = 0
  twos = 0
  while twos < n:
    digits += 1
    twos = 5*twos + 2
  t = n + twos
  rev = []
  while t > 0:
    d = t % 5
    if d == 0:
      rev.append('=')
    elif d == 1:
      rev.append('-')
    else:
      rev.append(str(d - 2))
    t //= 5
  return ''.join(reversed(rev))
```

```python
total = 0
for line in lines:
  total += to_num(line)
print(to_snafu(total))
```
