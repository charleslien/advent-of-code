```python
import numpy as np
rows = [[1] * 11]
def get_val(x, y):
  global rows
  while y >= len(rows):
    rows.append([1] + [0] * 7 + [1] * 3)
  return rows[y + 1][x + 1]

def set_val(x, y):
  global rows
  while y + 1 >= len(rows):
    rows.append([1] + [0] * 7 + [1] * 3)
  rows[y + 1][x + 1] = 1

def print_rows():
  global rows
  for row in reversed(rows):
    print(''.join('#' if x else '.' for x in row))

def mask(shape, x, y):
  global rows
  s = np.shape(shape)
  while y + s[0] >= len(rows):
    rows.append([1] + [0] * 7 + [1] * 3)
  return np.sum(np.multiply(np.array(rows[y+1:y+1+s[0]])[:,x+1:x+1+s[1]], shape))
```

```python
def get_height(num_shapes, period=False):
  global rows, max_ys, max_yi, shape_num
  rows = [[1] * 11]
  curr_index = 0
  shape_num = 1
  curr_x, curr_y = 2, 3
  max_y = -1
  max_ys = {(i, j): [] for i, _ in enumerate(raw) for j in range(5)}
  max_yi = {(i, j): [] for i, _ in enumerate(raw) for j in range(5)}
  i = 0
  while shape_num <= num_shapes or period:
    c = raw[i % len(raw)]
    d = -1 if c == '<' else 1
    curr_x += d
    if mask(shapes[curr_index], curr_x, curr_y):
      curr_x -= d
    curr_y -= 1
    if mask(shapes[curr_index], curr_x, curr_y):
      curr_y += 1
      for y, row in enumerate(shapes[curr_index]):
        for x, v in enumerate(row):
          if v:
            set_val(curr_x + x, curr_y + y)
            max_y = max(curr_y + y, max_y)
      if period:
        max_ys[i % len(raw), curr_index].append(max_y)
        max_yi[i % len(raw), curr_index].append(shape_num)

        curr_ys = max_ys[i % len(raw), curr_index]
        if len(curr_ys) >= 3:
          if curr_ys[-1] + curr_ys[-3] == curr_ys[-2] * 2:
            curr_yi = max_yi[i % len(raw), curr_index]
            return curr_yi[-1] - curr_yi[-2], curr_ys[-1] - curr_ys[-2]

      curr_index += 1
      curr_index %= 5
      shape_num += 1
      curr_x, curr_y = 2, max_y + 4
    i += 1
  return max_y+1
```
