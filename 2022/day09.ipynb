```python
lines = raw.splitlines()
```

```python
import re
p = re.compile(r'(R|L|D|U) (\d+)')
xh, yh = 0, 0
xt, yt = 0, 0
def move(d):
  global xh, yh, xt, yt
  if d == 'R':
    xh += 1
  if d == 'L':
    xh -= 1
  if d == 'U':
    yh += 1
  if d == 'D':
    yh -= 1

  if abs(xh - xt) > 1 or abs(yh - yt) > 1:
    if xt < xh:
      xt += 1
    elif xt > xh:
      xt -= 1

    if yt < yh:
      yt += 1
    elif yt > yh:
      yt -= 1
  return xt, yt
```

```python
pos = set()
for line in lines:
  d, l = p.match(line).groups()
  l = int(l)
  for _ in range(l):
    pos.add(move(d))
len(pos)
```

```python
rope = [[0, 0] for _ in range(10)]
def move2(d):
  global rope
  if d == 'R':
    rope[0][0] += 1
  if d == 'L':
    rope[0][0] -= 1
  if d == 'U':
    rope[0][1] += 1
  if d == 'D':
    rope[0][1] -= 1

  for i, h in enumerate(rope):
    if i == len(rope) - 1:
      break
    t = rope[i+1]

    if abs(h[0] - t[0]) > 1 or abs(h[1] - t[1]) > 1:
      if t[0] < h[0]:
        t[0] += 1
      elif t[0] > h[0]:
        t[0] -= 1

      if t[1] < h[1]:
        t[1] += 1
      elif t[1] > h[1]:
        t[1] -= 1
  return tuple(rope[-1])
```

```python
pos = set()
for line in lines:
  d, l = p.match(line).groups()
  l = int(l)
  for _ in range(l):
    pos.add(move2(d))
len(pos)
```
