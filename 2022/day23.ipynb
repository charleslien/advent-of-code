```python
lines = raw.splitlines()
```

```python
N = (-1, 0)
S = (1, 0)
E = (0, 1)
W = (0, -1)
```

```python
perpendicular = {
    N: E,
    S: E,
    W: N,
    E: N,
}
```

```python
elves = set()
for i, line in enumerate(lines):
  for j, c in enumerate(line):
    if c == '#':
      elves.add((i, j))
```

```python
dirs = [N, S, W, E]
for _ in range(10):
  source = {}
  blocked = set()
  for i, j in elves:
    move = False
    for di in [-1, 0, 1]:
      for dj in [-1, 0, 1]:
        if (di, dj) == (0, 0):
          continue
        if (i + di, j + dj) in elves:
          move = True
          break
    propd = (0, 0)
    if move:
      for d in dirs:
        di, dj = perpendicular[d]
        skip = False
        for dx in [-1, 0, 1]:
          if (i + d[0] + di * dx, j + d[1] + dj * dx) in elves:
            skip = True
            break
        if skip:
          continue
        propd = d
        break
    dest = i + propd[0], j + propd[1]
    if dest in source:
      blocked.add((i, j))
      blocked.add(source[dest])
    else:
      source[dest] = i, j
  new_elves = set()
  for elf in blocked:
    new_elves.add(elf)
  for dest, start in source.items():
    if start not in blocked:
      new_elves.add(dest)
  elves = new_elves
  dirs = dirs[1:] + [dirs[0]]
```
