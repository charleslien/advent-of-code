```python
curr = None
for i, line in enumerate(lines):
  if curr is not None:
    break
  for j, c in enumerate(line):
    if c == '^':
      curr = np.array([i, j])
      break
d = UP
visited = set([tuple(curr)])
while valid_index(lines, curr):
  while True:
    curr += d
    if not valid_index(lines, curr):
      break
    if get(lines, curr, '') == '#':
      curr -= d
      break
    visited.add(tuple(curr))
  d = np.array([d[1], -d[0]])
len(visited)
```

```python
start = None
for i, line in enumerate(lines):
  if start is not None:
    break
  for j, c in enumerate(line):
    if c == '^':
      start = np.array([i, j])
      break
grid = [list(line) for line in lines]
def has_loop(grid, start):
  curr = np.array(start)
  visited = {x: set() for x in [(0, 1), (0, -1), (1, 0), (-1, 0)]}
  d = np.array(UP)
  while valid_index(grid, curr):
    while True:
      curr += d
      if tuple(curr) in visited[tuple(d)]:
        return True
      visited[tuple(d)].add(tuple(curr))
      if not valid_index(grid, curr):
        break
      if get(grid, curr, '') == '#':
        curr -= d
        break
    d = np.array([d[1], -d[0]])
  return False

obstructions = set()
for i, j in visited:
  if grid[i][j] != '.':
    continue
  grid[i][j] = '#'
  if has_loop(grid, start):
    obstructions.add((i, j))
  grid[i][j] = '.'
len(obstructions)
```
