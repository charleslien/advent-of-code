```python
p = re.compile(r'p=(-?\d+),(-?\d+) v=(-?\d+),(-?\d+)')
```

```python
def find_pos(px, py, vx, vy, time):
  pos = (px + vx * time) % maxx, (py + vy * time) % maxy
  if pos[0] < maxx // 2 and pos[1] < maxy // 2:
    return 0
  if pos[0] > maxx // 2 and pos[1] < maxy // 2:
    return 1
  if pos[0] < maxx // 2 and pos[1] > maxy // 2:
    return 2
  if pos[0] > maxx // 2 and pos[1] > maxy // 2:
    return 3
maxx = 101
maxy = 103
quadrants = [0] * 4
for line in lines:
  if m := p.match(line):
    px, py, vx, vy = map(int, m.groups())
    if (pos := find_pos(px, py, vx, vy, 100)) is not None:
      quadrants[pos] += 1
np.prod(quadrants)
```

```python
def display_grid(points):
  for x in range(101):
    for y in range(103):
      if (x, y) in points:
        print('X', end='')
      else:
        print(' ', end='')
    print()
vals = [tuple(map(int, p.match(line).groups())) for line in lines]
t = 0
while True:
  poss = {((px + vx * t) % maxx, (py + vy * t) % maxy) for (px, py, vx, vy) in vals}
  if len(poss) == len(lines):
    display_grid(poss)
    break
  t += 1
```
