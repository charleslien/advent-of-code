```python
grid = np.array([list(line) for line in lines])
```

```python
for i, row in enumerate(grid):
  for j, c in enumerate(row):
    if c == 'S':
      start = (i, j)
    if c == 'E':
      end = (i, j)
heap = [(0, tuple(E), start)]
visited = set()

global_mins = {} # added for part 2
while heap:
  cost, d, pos = heapq.heappop(heap)
  if pos == end:
    print(cost)
    break
  if (d, pos) in visited:
    continue
  visited.add((d, pos))
  global_mins[(d, pos)] = cost # added for part 2
  if grid[tuple(np.array(pos) + d)] in '.E':
    heapq.heappush(heap, (cost + 1, d, tuple(np.array(pos) + d)))
  heapq.heappush(heap, (cost + 1000, tuple(np.array([[0, -1], [1, 0]]) @ d), pos))
  heapq.heappush(heap, (cost + 1000, tuple(np.array([[0, 1], [-1, 0]]) @ d), pos))
```

```python
visited = set()
chairs = set()
max_cost = cost
def dfs(turns, dist, d, pos):
  if turns > max_cost // 1000 or dist > max_cost % 1000:
    return
  if pos == end:
    global chairs
    chairs |= {v[1] for v in visited}
    return
  if (d, pos) in visited:
    return
  curr_global_min = global_mins.get((d, pos), float('inf'))
  curr_cost = turns * 1000 + dist
  if curr_cost > curr_global_min:
    return
  elif curr_cost < curr_global_min:
    global_mins[(d, pos)] = curr_cost
  visited.add((d, pos))
  if grid[pos[0] + d[0], pos[1] + d[1]] in '.E':
    dfs(turns, dist + 1, d, (pos[0] + d[0], pos[1] + d[1]))
  dfs(turns + 1, dist, (-d[1], d[0]), pos)
  dfs(turns + 1, dist, (d[1], -d[0]), pos)
  visited.remove((d, pos))
dfs(0, 0, tuple(E), start)
print(len(chairs) + 1)
```
