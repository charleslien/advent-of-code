```python
bricks = []
for i, line in enumerate(lines):
  a, b = line.split('~')
  a = [int(x) for x in a.split(',')]
  b = [int(x) for x in b.split(',')]
  brick = (
      min(a[0], b[0]),
      max(a[0], b[0]),
      min(a[1], b[1]),
      max(a[1], b[1]),
      min(a[2], b[2]),
      max(a[2], b[2]),
  )
  bricks.append(brick)
bricks.sort(key=lambda x: min(x[-1], x[-2]))
```

```python
def collide(b1, b2):
  while b1:
    if (min(b1[0], b1[1]) > max(b2[0], b2[1])) or (min(b2[0], b2[1]) > max(b1[0], b1[1])):
      return False
    b1 = b1[2:]
    b2 = b2[2:]
  return True

def supports(b1, b2):
  return collide(b1, (b2[0], b2[1], b2[2], b2[3], b2[4] - 1, b2[5] - 1))

def settle(bricks, reset_z=True):
  n = list(bricks)
  while True:
    moved = False
    max_z = 0
    for i in range(len(n)):
      new_brick = n[i]
      if reset_z and new_brick[4] > max_z and new_brick[5] > max_z:
        new_brick = new_brick[0], new_brick[1], new_brick[2], new_brick[3], max_z + 1, new_brick[5] - new_brick[4] + max_z + 1
      while not any(collide(new_brick, n[j]) for j in range(len(n)) if i != j) and new_brick[4] > 0 and new_brick[5] > 0:
        new_brick = new_brick[0], new_brick[1], new_brick[2], new_brick[3], new_brick[4] - 1, new_brick[5] - 1
      new_brick = new_brick[0], new_brick[1], new_brick[2], new_brick[3], new_brick[4] + 1, new_brick[5] + 1
      if n[i] != new_brick:
        n[i] = new_brick
        moved = True
      max_z = max([max_z, new_brick[5]])
    if not moved:
      return n
```

```python
settled = settle(bricks)
```

```python
sup = []
supd = [[] for _ in settled]
for i, bi in enumerate(settled):
  sup.append([])
  for j, bj in enumerate(settled):
    if i == j:
      continue
    if supports(bi, bj):
      sup[i].append(j)
      supd[j].append(i)
```
