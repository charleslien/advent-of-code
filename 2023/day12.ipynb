```python
def get_num_completions(spring, nums):
  if '?' not in spring:
    lens = [len(x) for x in spring.split('.') if x]
    if lens == nums:
      return 1
    return 0
  first_unk = spring.find('?')
  return get_num_completions(spring[:first_unk] + '.' + spring[first_unk+1:], nums) + get_num_completions(spring[:first_unk] + '#' + spring[first_unk+1:], nums)
```

```python
def get_num_completions(spring, lo, nums, index, memo=None):
  if (lo, index) in memo:
    return memo[lo, index]
  c = nums[index]
  if '.' in spring[lo:lo+c]:
    memo[lo, index] = 0
    return 0
  if lo+c > len(spring):
    memo[lo, index] = 0
    return 0
  if lo+c < len(spring) and spring[lo+c] == '#':
    memo[lo, index] = 0
    return 0
  if index == len(nums) - 1:
    if '#' in spring[lo+c:]:
      memo[lo, index] = 0
      return 0
    memo[lo, index] = 1
    return 1
  total = 0
  for i in range(lo+c+1, len(spring)):
    total += get_num_completions(spring, i, nums, index+1, memo=memo)
    if spring[i] == '#':
      break
  memo[lo, index] = total
  return total
```
