```python
from collections import deque
def max_flow_le3(u, v):
  c_flow = 0
  flow = set()
  while True:
    p = {u: None}
    q = deque([u])
    while q:
      c = q.popleft()
      for n in edges[c]:
        if (c, n) in flow:
          continue
        if n in p:
          continue
        p[n] = c
        if n == v:
          break
        q.append(n)
      if n == v:
        break
      n = None
    if n is None:
      return flow
    while p[c] is not None:
      if (c, p[c]) in flow:
        flow.remove((c, p[c]))
      else:
        flow.add((p[c], c))
      c = p[c]
    c_flow += 1
    if c_flow > 3:
      return False
```
