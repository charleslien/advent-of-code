```python
def get_corresponding(n, block):
  for sink, src, l in block:
    if n >= src and n < src + l:
      return sink + n - src
  return n
```

```python
def get_corresponding_range(r, block):
  low, high = r
  ret = []
  ignore = []
  for sink, src, l in block:
    if low < src + l and high > src:
      ret.append((max(low, src) + sink - src, min(high, src + l) + sink - src))
      ignore.append((max(low, src), min(high, src+l)))
  prev = low
  for l, h in sorted(ignore):
    ret.append((prev, l))
    prev = h
  ret.append((prev, high))
  return [x for x in ret if x[0] < x[1]]
```
