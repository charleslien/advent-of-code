```python
stack = [(None, start)]
path = []
visited = set()
most = 0
while stack:
  _, curr = stack.pop()
  i, j = curr
  path.append(curr)
  visited.add(curr)
  if curr == end:
    most = max(most, len(path))
    while stack and stack[-1][0] != path[-1]:
      c = path.pop()
      visited.remove(c)
    continue
  added = False
  for di, dj in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
    n = i + di, j + dj
    if n[0] < 0 or n[0] >= len(lines) or n[1] < 0 or n[1] > len(lines[0]):
      continue
    if n in visited:
      continue
    c = lines[n[0]][n[1]]
    if c == '#':
      continue
    if di == -1 and c == 'v':
      continue
    if di == 1 and c == '^':
      continue
    if dj == -1 and c == '>':
      continue
    if dj == 1 and c == '<':
      continue
    added = True
    stack.append((curr, n))
  if not added:
    while stack and stack[-1][0] != path[-1]:
      c = path.pop()
      visited.remove(c)
most - 1
```

```python
intersections = {start, end}
for i, row in enumerate(lines):
  for j, c in enumerate(row):
    if c == '#':
      continue
    count = 0
    for di, dj in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
      ni = i + di
      nj = j + dj
      if ni >= 0 and ni < len(lines) and nj >= 0 and nj < len(lines[0]) and lines[ni][nj] != '#':
        count += 1
    if count > 2:
      intersections.add((i, j))
```

```python
from collections import deque
edges = {}
for int1 in intersections:
  edges[int1] = []
  q = deque([(int1, 0)])
  visited = set()
  while q:
    curr, dist = q.popleft()
    i, j = curr
    visited.add(curr)
    for di, dj in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
      ni = i + di
      nj = j + dj
      if ni < 0 or ni >= len(lines) or nj < 0 or nj >= len(lines[0]) or lines[ni][nj] == '#':
        continue
      if (ni, nj) in visited:
        continue
      visited.add((ni, nj))
      if (ni, nj) in intersections:
        edges[int1].append(((ni, nj), dist + 1))
        continue
      q.append(((ni, nj), dist + 1))
```

```python
for edgelist in edges.values():
  edgelist.sort(key=lambda x: -x[1])
```

```python
edge_weights = {}
for u, vs in edges.items():
  for v, d in vs:
    edge_weights[u, v] = d
```

```python
def pathlen(path):
  total = 0
  for i in range(1, len(path)):
    for t, d in edges[path[i-1]]:
      if t == path[i]:
        total += d
        break
  return total
```

```python
def calc_ub(visited):
  remaining = len(intersections) - len(visited)
  edgelens = []
  for k, v in edges.items():
    if k in visited:
      continue
    edgelens.append(v[0][1])
  edgelens.sort()
  return sum(edgelens[-remaining:])
```

```python
def calc_ub_2(visited, curr):
  remaining = intersections - visited
  mst = {curr}
  mst_weight = 0
  while remaining:
    max_v = None
    max_w = 0
    for v in remaining:
      for u in mst:
        w = edge_weights.get((u, v), 0)
        if w > max_w:
          max_w = w
          max_v = v
    if max_v is None:
      if end not in mst:
        return 0
      return mst_weight
    mst_weight += max_w
    remaining.remove(max_v)
    mst.add(max_v)
  return mst_weight
```
