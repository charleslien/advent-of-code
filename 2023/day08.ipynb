```python
import re
p = re.compile(r'(\w\w\w) = \((\w\w\w), (\w\w\w)\)')
```

```python
d = {}
for line in lines:
  a, b, c = p.match(line).groups()
  d[a] = {'L': b, 'R': c}
```

```python
count = 0
curr = 'AAA'
while curr != 'ZZZ':
  for inst in lr:
    if curr == 'ZZZ':
      break
    curr = d[curr][inst]
    count += 1
count
```

```python
periods = []
offsets = []
s_offsets = []
for start in d:
  if not start.endswith('A'):
    continue
  visited = {(start, 0): len(lr) - 1}
  curr = start
  count = 0
  offsets.append([])
  exit = False
  while not exit:
    for i, inst in enumerate(lr):
      curr = d[curr][inst]
      count += 1
      if curr.endswith('Z'):
        offsets[-1].append(count)
      if (curr, i) in visited:
        periods.append(count - visited[curr, i])
        s_offsets.append(visited[curr, i])
        exit = True
        break
      visited[curr, i] = count
```

```python
# Importantly, there is only one offset per starting point and that start point is the same as the period.
# This means we can take the LCM of all the periods to get the answer.
print(f'{offsets=}')
print(f'{periods=}')
print(f'{s_offsets=}')
```

```python
import math
total = 1
for p in periods:
  total = total * p // math.gcd(total, p)
total
```
