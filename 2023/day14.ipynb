```python
def rolled(s):
  r = []
  rocks = 0
  for i, c in enumerate(s):
    if c == 'O':
      rocks += 1
    if c == '#':
      while len(r) < i - rocks:
        r.append('.')
      r.extend(['O'] * rocks)
      r.append('#')
      rocks = 0
  while len(r) < len(s) - rocks:
    r.append('.')
  r.extend(['O'] * rocks)

  return ''.join(r)
```

```python
cycles = {''.join(lines): 0}
boards = [lines]
n = 0
curr = lines
while True:
  n += 1
  for _ in range(4):
    curr = [''.join(line[j] for line in reversed(curr)) for j in range(len(curr[0]))]
    curr = [rolled(line) for line in curr]
  boards.append(curr)
  x = ''.join(curr)
  if x in cycles:
    period = n - cycles[x]
    offset = cycles[x]
    break
  cycles[x] = n
```
