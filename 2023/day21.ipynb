```python
steps = 26501365
```

```python
def get_results(dupes):
  ls = []
  for i in range(-dupes, dupes + 1):
    for line in lines:
      if 'S' in line:
        if i == 0:
          ls.append('.' * 131 * dupes + line + '.' * 131 * dupes)
        else:
          ls.append('.' * 131 * (2 * dupes + 1))
        continue
      ls.append(line * (2 * dupes + 1))
  curr = set()
  visited = set()
  for i, line in enumerate(ls):
    if curr:
      break
    for j, c in enumerate(line):
      if c == 'S':
        curr.add((i, j))
        visited.add((i, j))
        break
  for _ in range(steps % 131 + 131 * dupes):
    n = set()
    for i, j in curr:
      for di, dj in [(-1, 0), (1, 0), (0, 1), (0, -1)]:
        if (i + di, j + dj) in visited:
          continue
        if i + di < len(ls) and i + di >= 0 and j + dj < len(ls[0]) and j + dj >= 0 and ls[i + di][j + dj] != '#':
          n.add((i + di, j + dj))
          visited.add((i + di, j + dj))
    curr = n
  return len([(i, j) for (i, j) in visited if (i + j + dupes) % 2 == 1])
```

```python
r0 = get_results(0)
r2 = get_results(2)
r4 = get_results(4)
```

```python
from scipy.interpolate import lagrange
x = [0, 2, 4]
y = [r0, r2, r4]
p = lagrange(x, y)
```
