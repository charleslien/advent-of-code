```python
def get_num_energized(i, j, di, dj):
  en = set()
  visited = set()
  stack = [(i, j, di, dj)]
  while stack:
    curr = stack.pop()
    i, j, di, dj = curr
    if curr in visited:
      continue
    if i < 0 or i >= len(lines) or j < 0 or j >= len(lines[0]):
      continue
    visited.add(curr)
    en.add((i, j))
    c = lines[i][j]
    if c == '.':
      stack.append((i+di, j+dj, di, dj))
      continue
    if c == '/':
      di, dj = -dj, -di
      stack.append((i+di, j+dj, di, dj))
      continue
    if c == '\\':
      di, dj = dj, di
      stack.append((i+di, j+dj, di, dj))
      continue
    if c == '-':
      if di == 0:
        stack.append((i+di, j+dj, di, dj))
        continue
      stack.append((i, j+1, 0, 1))
      stack.append((i, j-1, 0, -1))
      continue
    if dj == 0:
      stack.append((i+di, j+dj, di, dj))
      continue
    stack.append((i+1, j, 1, 0))
    stack.append((i-1, j, -1, 0))
  return len(en)
```

```python
m = 0
for si in range(len(lines)):
  m = max(m, get_num_energized(si, 0, 0, 1))
  m = max(m, get_num_energized(si, len(lines[0]) - 1, 0, -1))
for sj in range(len(lines[0])):
  m = max(m, get_num_energized(0, sj, 1, 0))
  m = max(m, get_num_energized(len(lines) - 1, sj, -1, 0))
m
```
