```python
val = {c: i for i, c in enumerate('AKQJT98765432')}
def power(h):
  counts = {}
  for c in h:
    counts[c] = counts.get(c, 0) + 1
  v = {}
  for x in counts.values():
    v[x] = v.get(x, 0) + 1
  if v == {5: 1}:
    return 0
  if v == {4: 1, 1: 1}:
    return 1
  if v == {2: 1, 3: 1}:
    return 2
  if v == {3: 1, 1: 2}:
    return 3
  if v == {2: 2, 1: 1}:
    return 4
  if v == {2: 1, 1: 3}:
    return 5
  if v == {1: 5}:
    return 6
def tieb(h):
  return tuple(val[c] for c in h)
```

```python
def j_select(h):
  counts = {}
  for c in h:
    counts[c] = counts.get(c, 0) + 1
  for k, v in sorted(counts.items(), key=lambda x: -x[1]):
    if k == 'J':
      continue
    return k
  return 'A'
```

```python
def new_j_hand(h):
  js = j_select(h)
  return ''.join(c if c != 'J' else js for c in h)
```

```python
snjs = list(reversed(sorted(vs, key=lambda x: (power(new_j_hand(x[0])), tieb(x[0])))))
```
