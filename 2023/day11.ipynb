```python
lines = raw.splitlines()
```

```python
n = len(lines)
m = len(lines[0])
```

```python
g = set()
for i in range(n):
  for j in range(m):
    if lines[i][j] == '#':
      g.add((i, j))
```

```python
erows = [i for i in range(n) if '#' not in lines[i]]
ecols = [j for j in range(m) if all(line[j] != '#' for line in lines)]
```

```python
total = 0
for i1, j1 in g:
  for i2, j2 in g:
    di = abs(i1 - i2)
    dj = abs(j1 - j2)
    total += di + dj
    for x in erows:
      if (min(i1, i2) < x and max(i1, i2) > x):
        total += 1
    for x in ecols:
      if (min(j1, j2) < x and max(j1, j2) > x):
        total += 1
total // 2
```
