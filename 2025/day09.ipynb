```python
ps = [[int(x) for x in line.split(',')] for line in blocks]
m = 0
for i, u in enumerate(ps):
  u = np.array(u)
  for j in range(i + 1, len(ps)):
    v = ps[j]
    m = max(m, np.prod(np.abs(u - v) + [1, 1]))
int(m)
```

```python
cols = set()
rows = set()
for x, y in ps:
  cols.add(x)
  rows.add(y)
cols = list(sorted(cols))
rows = list(sorted(rows))

new_ps = [(cols.index(p[0]), rows.index(p[1])) for p in ps]
grid = np.zeros((len(cols) + 1, len(rows) + 1), dtype=np.int32)

prev_p = None
for p in new_ps:
  if prev_p is None:
    prev_p = p
    continue
  if prev_p[0] == p[0]:
    for i in range(min(prev_p[1], p[1]), max(prev_p[1], p[1]) + 1):
      grid[prev_p[0], i] = 1
  else:
    for i in range(min(prev_p[0], p[0]), max(prev_p[0], p[0]) + 1):
      grid[i, prev_p[1]] = 1
  prev_p = p

stack = []
for i in range(len(grid)):
  stack.append((i, 0))
  stack.append((i, len(grid[0]) - 1))
for j in range(len(grid[0])):
  stack.append((0, j))
  stack.append((len(grid) - 1, j))
while stack:
  c = stack.pop()
  if get(grid, c, 2):
    continue
  grid[c] = 2
  for d in ORTHOGONAL_DIRS:
    stack.append(tuple(d + c))

new_ps = np.array(new_ps)
m = 0
for i, u in enumerate(new_ps):
  for j in range(i + 1, len(new_ps)):
    v = new_ps[j]
    x0 = cols[min(u[0], v[0])]
    x1 = cols[max(u[0], v[0])]
    y0 = rows[min(u[1], v[1])]
    y1 = rows[max(u[1], v[1])]
    area = (x1 - x0 + 1) * (y1 - y0 + 1)
    if area <= m:
      continue
    invalid = False
    for ip in range(min(u[0], v[0]), max(u[0], v[0]) + 1):
      if invalid:
        break
      for jp in range(min(u[1], v[1]), max(u[1], v[1]) + 1):
        if invalid:
          break
        if grid[ip, jp] == 2:
          invalid = True
          break
    if not invalid:
      m = area
int(m)
```
