```python
ins = {}
for line in blocks:
  head, *rest = line.split()
  u = head[:-1]
  for out in rest:
    ins[out] = ins.get(out, [])
    ins[out].append(u)

@cache
def n_paths(node):
  if node == 'you':
    return 1
  if node not in ins:
    return 0
  return sum(n_paths(n) for n in ins[node])

n_paths('out')
```

```python
@cache
def n_paths_with_req(start, target):
  if target == start:
    return 1
  if target not in ins:
    return 0
  return sum(n_paths_with_req(start, n) for n in ins[target])

fd = n_paths_with_req('svr', 'fft')
fd *= n_paths_with_req('fft', 'dac')
fd *= n_paths_with_req('dac', 'out')

df = n_paths_with_req('svr', 'dac')
df *= n_paths_with_req('dac', 'fft')
df *= n_paths_with_req('fft', 'out')

fd + df
```
