```python
raw = '''503950-597501,73731-100184,79705998-79873916,2927-3723,35155-50130,52-82,1139-1671,4338572-4506716,1991-2782,1314489-1387708,8810810-8984381,762581-829383,214957-358445,9947038-10058264,4848455367-4848568745,615004-637022,5827946-5911222,840544-1026063,19-46,372804-419902,486-681,815-1117,3928-5400,28219352-28336512,6200009-6404247,174-261,151131150-151188124,19323-26217,429923-458519,5151467682-5151580012,9354640427-9354772901,262-475,100251-151187,5407-9794,8484808500-8484902312,86-129,2-18'''
```

```python
blocks = [[int(x) for x in row.split('-')] for row in raw.split(',')]
```

```python
def double(n):
  return int(str(n) + str(n))
count = 0
for lo, hi in blocks:
  n = len(str(lo))
  if n % 2 == 0:
    pre = str(lo)[:n // 2]
  else:
    pre = '1' + '0' * (n // 2)
  pre = int(pre)
  if double(pre) < lo:
    pre += 1

  n = len(str(hi))
  if n % 2 == 0:
    pre2 = str(hi)[:n // 2]
  else:
    pre2 = '9' * (n // 2)
  pre2 = int(pre2)
  if double(pre2) > hi:
    pre2 -= 1

  for x in range(pre, pre2 + 1):
    count += int(str(x) + str(x))
count
```

```python
counts = set()
def repeat(num, k):
  return int(str(num) * k)

for lo, hi in blocks:
  hin = len(str(hi))
  lon = len(str(lo))
  for repeats in range(2, hin + 1):
    if hin % repeats == 0:
      c = int(str(hi)[:hin // repeats])
    else:
      c = int(('9' * hin)[:hin // repeats])
    if repeat(c, repeats) > hi:
      c -= 1
    while repeat(c, repeats) >= lo:
      counts.add(repeat(c, repeats))
      c -= 1
sum(counts)
```
