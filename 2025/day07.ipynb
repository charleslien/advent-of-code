```python
blocks = process_raw(raw)
```

```python
s = set()
for i, v in enumerate(blocks[0]):
  if v == 'S':
    s.add(i)
    break

count = 0
for line in blocks:
  new_set = set()
  for i in s:
    c = line[i]
    if c == '^':
      new_set.add(i - 1)
      new_set.add(i + 1)
      count += 1
    else:
      new_set.add(i)
  s = new_set
count
```

```python
counts = {}
for i, v in enumerate(blocks[0]):
  if v == 'S':
    counts[i] = 1
    break

for line in blocks:
  if '^' not in line:
    continue
  new_counts = {}
  for i, count in counts.items():
    c = line[i]
    if c == '^':
      new_counts[i - 1] = new_counts.get(i - 1, 0) + count
      new_counts[i + 1] = new_counts.get(i + 1, 0) + count
    else:
      new_counts[i] = new_counts.get(i, 0) + count
  counts = new_counts
sum(counts.values())
```
